Changing the date column for further analysis

```python
import calendar
df2[' Date'] = pd.to_datetime(df2[' Date'], dayfirst=True)
df2[' Frequency'] = df2[' Frequency'].astype('category')
df2['Month'] = df2[' Date'].dt.month
df2['Month_int'] = df2['Month'].apply(lambda x: int(x))
df2['Month_name'] = df2['Month_int'].apply(lambda x: calendar.month_abbr[x])
df2['Region'] = df2['Region'].astype('category')
df2.drop(columns='Month', inplace=True)
df2.head(11)
```

```python
# Select only the numerical columns
numerical_columns = [' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)', 'longitude', 'latitude', 'Month_int']

# Create a pair plot
sns.pairplot(df2[numerical_columns], markers=["o", "s", "D"])
plt.show()
```

```python
region_stats = df2.groupby('Region').mean([({' Estimated Unemployment Rate (%)',' Estimated Employed',' Estimated Labour Participation Rate (%)'})])
# Display the computed statistics
region_stats
```

Visualising the average Estimated Unemployment Rate

```python
# Select columns of interest
plot_ump = df2[[' Estimated Unemployment Rate (%)', 'Region']]

# Group by ' Region' and compute the mean
df_unemp = plot_ump.groupby('Region').mean().reset_index()

# Sort the DataFrame by ' Estimated Unemployment Rate (%)'
df_unemp = df_unemp.sort_values(' Estimated Unemployment Rate (%)')

# Set the Seaborn style
sns.set(style="whitegrid")

# Create a bar plot using Seaborn with color-coded bars
plt.figure(figsize=(10, 6))
sns.barplot(data=df_unemp, x='Region', y=' Estimated Unemployment Rate (%)', palette='Set3')
plt.title('Average Unemployment Rate in Each Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```
