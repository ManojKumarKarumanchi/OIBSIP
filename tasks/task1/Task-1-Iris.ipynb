We'll split the dataset into training and testing sets and then scale the features.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# input data
X = iris.drop(columns=['Species'])
# output data
y = iris['Species']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Now using the KNN algorithm will calculate the accuracy score of this model

```python
# knn - k-nearest neighbours
## Training the model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)

# print metric to get performance
print("Accuracy: ",model.score(X_test, y_test) * 100)
```

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred
```

Now letâ€™s input a set of measurements of the iris flower and use the model to predict the iris species:

```python
x_new = np.array([[5, 2.9, 1, 0.2, 3]])
prediction = model.predict(x_new)
print("Prediction: {}".format(prediction))
```
